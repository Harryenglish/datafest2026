```python
dept_clean = departments.copy()

dept_clean.columns = dept_clean.columns.str.strip()

dept_clean = dept_clean.drop_duplicates(subset=["DepartmentKey"])

dept_clean["City"] = (
    dept_clean["City"]
    .astype(str)
    .str.strip()
    .str.title()
)
```
